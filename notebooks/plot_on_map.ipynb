```python
#First we plot dots for every logged data gnss coord
import pandas as pd
from mapboxgl.utils import *
from mapboxgl.viz import *

#(you should setup your token on the shell environment you ran the notebook from)
#(token is available on your mapbox account, starts with "pk.")
token = os.getenv('MAPBOX_ACCESS_TOKEN')

#don't let pandas truncate the outputs, we need them to analyze data/coords
pd.set_option("display.max_rows", None, "display.max_columns", None)

#load cols: [0]lat, [1]lon, [4]alt and [5]bea
np_gnss = np.load(example_segment + 'processed_log/GNSS/live_gnss_ublox/value')[:,[0,1,4,5]];

driven_df = pd.DataFrame(data=np_gnss);

#we need 7 decimal places for sub meter precision
#https://en.wikipedia.org/wiki/Decimal_degrees
pd.options.display.float_format = '{:.10f}'.format

driven_df.columns = ['lat', 'lon', 'alt', 'bea']
driven_df['tag'] = 'driven'
#driven segment dataframe is ready
driven_df
```

```python
#NEED TO GRAB AND NORMALIZE (TO FLOAT) DATA FROM BOTH DATAFRAMES, THEN PRINT IT TO MAP
#map_df will be the dataframe used to draw on map
map_df = pd.DataFrame(columns = ['lat', 'lon', 'tag'])
map_df = pd.concat([driven_df])

map_gjson = df_to_geojson(map_df, lat='lat', lon='lon')
viz = CircleViz(map_gjson, 
                access_token=token, 
                radius=1, 
                center=(map_df.lon.mean(), map_df.lat.mean()), #center the map on dataframe
                zoom=13,
                style='mapbox://styles/mapbox/satellite-v9')

viz.color_property = 'tag'
viz.default_color = 'blue'
viz.color_function_type = 'match'
viz.color_stops = [['osm_nodes', 'rgb(211,25,25)'],['driven', 'rgb(25,211,25)'],['osm_bbox','rgb(25,25,211)'],['pink','rgb(255,192,203)']]

viz.show()
```

```python
#is geopy2 still really maintained? reliable?
from geopy import Point
from geopy.distance import geodesic

#grab driven data gnss inicial coords
lat_start = driven_df['lat'][0]
lon_start = driven_df['lon'][0]
bea_start = driven_df['bea'][0]
print('center', lat_start, lon_start, bea_start)

# cali hw lane width = 3.6m
dist = 10/1000 #meters

#use geopy to compute the coords based on bearing and distance
#use gnss bearing +-90deg to plot dots perpendicular to bearing (y no say dis) useful to check lane width
osm_bbox_nw = geodesic(kilometers=dist).destination(Point(lat_start, lon_start), 315) #northwest
osm_bbox_se = geodesic(kilometers=dist).destination(Point(lat_start, lon_start), 135) #southeast

#we only need two points for the osm overpass query, but could use all four vertices to represent the
#osm overpass bounding box nicely on the map
#also: more points may be useful later on to comply with overpass API calls
#https://wiki.openstreetmap.org/wiki/Overpass_API/Overpass_QL#Global_bounding_box_.28bbox.29
#[...]in order: southern-most latitude, western-most longitude, northern-most latitude, eastern-most longitude.
#swapping lats and lons between the two initial coords to get all the four bounding box corners
#may be negligible for small distances but will start losing precision with longer distances
osm_bbox_ne = geodesic(kilometers=dist).destination(Point(lat_start, lon_start), 45) #northeast
osm_bbox_sw = geodesic(kilometers=dist).destination(Point(lat_start, lon_start), 225) #southwest

osm_bbox_df = pd.DataFrame([[osm_bbox_nw[0], osm_bbox_nw[1],'osm_bbox_nw'],
                            [osm_bbox_se[0], osm_bbox_se[1],'osm_bbox_se'],
                            [osm_bbox_ne[0], osm_bbox_ne[1],'osm_bbox_ne'],
                            [osm_bbox_sw[0], osm_bbox_sw[1],'osm_bbox_sw']
                           ],
                           columns = ['lat', 'lon', 'id']
                          )
osm_bbox_df['tag'] = 'osm_bbox'

osm_bbox_df
```

```python

map_df = pd.concat([driven_df,osm_bbox_df])

map_gjson = df_to_geojson(map_df, lat='lat', lon='lon')
viz = CircleViz(map_gjson, 
                access_token=token, 
                radius=1, 
                center=(map_df.lon.mean(), map_df.lat.mean()), #center the map on driven path
                zoom=13,
                style='mapbox://styles/mapbox/satellite-v9')

viz.color_property = 'tag'
viz.default_color = 'blue'
viz.color_function_type = 'match'
viz.color_stops = [['osm_nodes', 'rgb(211,25,25)'],['driven', 'rgb(25,211,25)'],['osm_bbox','rgb(25,25,211)'],['pink','rgb(255,192,203)']]

viz.show()
```

```python
import overpy

#from Overpass API docs:
#southern-most latitude, western-most longitude, northern-most latitude, eastern-most longitude
querystr="""[out:json][timeout:10];
(
   way({sm}, {wm}, {nm}, {em});
);
._;
>;
out body;
""".format(sm=min(osm_bbox_df['lat']), #south-most
           wm=min(osm_bbox_df['lon']), #west-most
           nm=max(osm_bbox_df['lat']), #north-most
           em=max(osm_bbox_df['lon'])  #east-most
          )

# print(querystr)

api = overpy.Overpass()

result = api.query(querystr)
```

```python
nodes = result.get_nodes();

print(len(nodes));
```

```python
#create a dataframe from osm nodes data
waynodes_df = pd.DataFrame([[node.id, float(node.lat), float(node.lon)] for node in nodes], columns=['id','lat','lon'])

waynodes_df['tag']='osm_nodes'

# waynodes_df
```

```python
map_df = pd.concat([driven_df,osm_bbox_df,waynodes_df])

map_gjson = df_to_geojson(map_df, lat='lat', lon='lon')
viz = CircleViz(map_gjson, 
                access_token=token, 
                radius=1, 
                center=(map_df.lon.mean(), map_df.lat.mean()), #center the map on driven path
                zoom=13,
                style='mapbox://styles/mapbox/satellite-v9')

viz.color_property = 'tag'
viz.default_color = 'blue'
viz.color_function_type = 'match'
viz.color_stops = [['osm_nodes', 'rgb(211,25,25)'],['driven', 'rgb(25,211,25)'],['osm_bbox','rgb(25,25,211)']]

viz.show()
```

```python
# We can project the path driven onto the first image

# first we convert the frame_positions to the frame
# defined by the pose of the first frame
ecef_from_local = orient.rot_from_quat(frame_orientations[0])
local_from_ecef = ecef_from_local.T
frame_positions_local = np.einsum('ij,kj->ki', local_from_ecef, frame_positions - frame_positions[0])


from lib.camera import img_from_device, denormalize, view_frame_from_device_frame
def draw_path(device_path, img, width=1, height=1.2, fill_color=(128,0,255), line_color=(0,255,0)):
  device_path_l = device_path + np.array([0, 0, height])                                                                    
  device_path_r = device_path + np.array([0, 0, height])                                                                    
  device_path_l[:,1] -= width                                                                                               
  device_path_r[:,1] += width

  img_points_norm_l = img_from_device(device_path_l)
  img_points_norm_r = img_from_device(device_path_r)
  img_pts_l = denormalize(img_points_norm_l)
  img_pts_r = denormalize(img_points_norm_r)

  # filter out things rejected along the way
  valid = np.logical_and(np.isfinite(img_pts_l).all(axis=1), np.isfinite(img_pts_r).all(axis=1))
  img_pts_l = img_pts_l[valid].astype(int)
  img_pts_r = img_pts_r[valid].astype(int)

  for i in range(1, len(img_pts_l)):
    u1,v1,u2,v2 = np.append(img_pts_l[i-1], img_pts_r[i-1])
    u3,v3,u4,v4 = np.append(img_pts_l[i], img_pts_r[i])
    pts = np.array([[u1,v1],[u2,v2],[u4,v4],[u3,v3]], np.int32).reshape((-1,1,2))
    cv2.fillPoly(img,[pts],fill_color)
    cv2.polylines(img,[pts],True,line_color)


import cv2
img = imread(example_segment + 'preview.png')
draw_path(frame_positions_local[1:1000], img)
figsize(12,12);
imshow(img);
title('Driven path projected onto first image', fontsize=25);
```
